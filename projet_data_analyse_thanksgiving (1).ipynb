```python
# Création d'un deuxième objet Series indiquant des booléens sur les valeurs d'une colonne
pumpkin_isnull = pd.isnull(data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin"])
```

## Convertir l'âge en une valeur numérique

```python
# Afficher les tranches d'âges de la colonne 'Âge'
data["Age"].value_counts()
```

```python
# Création d'une fonction
def extract_age(age_str):
    if pd.isnull(age_str):
        return None
    age_str = age_str.split(" ")[0]
    age_str = age_str.replace("+", "")
    return int(age_str)

# Création d'une nouvelle colonne dans notre DataFrame en prenant en compte les éléments de la fonction
data["int_age"] = data["Age"].apply(extract_age)

# Appliquer la méthode describe sur la colonne 'int_age' du DataFrame
data["int_age"].describe()
```

## Convertir les revenus en valeurs numériques

```python
# Afficher les différentes catégories d'une colonne
data["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts()
```

```python
# Création d'une fonction 
def extract_income(income_str):
    if pd.isnull(income_str):
        return None 
    income_str = income_str.split(" ")[0]
    if income_str == "Prefer":
        return None
    income_str = income_str.replace("$", "")
    income_str = income_str.replace(",", "")
    return int(income_str)

# Création d'une nouvelle colonne dans notre DataFrame en prenant en compte les éléments de la fonction
data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(extract_income)

# Appliquer la méthode describe sur la colonne 'int_income' du DataFrame
data["int_income"].describe()
```

## Corrélation entre distance et revenus

```python
# Filtrer les données du DataFrame 'int_income' par rapport à -150 000
less_150000 = data[data["int_income"] < 150000]

# Afficher les personnes qui gagnent moins de 150 000 $
less_150000["How far will you travel for Thanksgiving?"].value_counts()
```

```python
# Filtrer les données du DataFrame 'int_income' par rapport à +150 000
more_150000 = data[data["int_income"] > 150000]

# Afficher les personnes qui gagnent plus de 150 000 $
more_150000["How far will you travel for Thanksgiving?"].value_counts()
```

## Plats principaux les plus populaires

```python
import pandas as pd

# Affichage du DataFrame et spécifier dans les paramètres la fonction encoding pour coder correctement
data = pd.read_csv("thanksgiving.csv", encoding="latin-1")
data.head()

import matplotlib.pyplot as plt
from numpy import arange

# Compter les occurrences de chaque plat principal
plats = data['What is typically the main dish at your Thanksgiving dinner?'].value_counts()

bar_heights = plats.values
bar_positions = arange(len(plats)) + 1
tick_positions = range(1, len(plats) + 1)

# Création du graphique 
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(plats.index, rotation=45, ha='right')

# Définir les titres du graphique
ax.set_xlabel('Plat principal')
ax.set_ylabel('Nombre de réponses')
ax.set_title('Plats principaux servis lors de Thanksgiving')

# Affichage du graphique
plt.show()
```

## Répartition par genre (Hommes/Femmes)

```python
import matplotlib.pyplot as plt
from numpy import arange

genres = data['What is your gender?'].value_counts()

bar_heights = genres.values
bar_positions = arange(len(genres)) + 1
tick_positions = range(1, len(genres) + 1)

# Création du graphique 
fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(bar_positions, bar_heights, 0.5, color=['steelblue', 'salmon'])
ax.set_xticks(tick_positions)
ax.set_xticklabels(genres.index, rotation=0)

# Définir les titres du graphique
ax.set_xlabel('Genre')
ax.set_ylabel('Nombre de réponses')
ax.set_title('Répartition des sondés par genre')

# Affichage du graphique
plt.show()
```

## Graphique en courbe : L'évolution du nombre de réponses par tranche d'âge selon le genre (Homme/Femme) 

```python
# Définir les tranches d'âge dans l'ordre logique
ordre_ages = ['18 - 29', '30 - 44', '45 - 59', '60+']

# Définir les genres et les couleurs associées
genres = ['Male', 'Female']
colors = ['blue', 'red']

fig = plt.figure(figsize=(10, 6))

# Boucle pour tracer une courbe par genre
for i in range(len(genres)):
    # Filtrer les données pour le genre i
    subset = data[data['What is your gender?'] == genres[i]]
    
    # Compter le nombre de réponses par tranche d'âge
    counts = subset['Age'].value_counts().reindex(ordre_ages)
    
    # Tracer la courbe avec la couleur et le label correspondant
    label = genres[i]
    plt.plot(ordre_ages, counts.values, c=colors[i], label=label)

plt.legend(loc='upper left')

# Ajout d'un titre au graphique
plt.title("Répartition par tranche d'âge selon le genre")

# Ajout des intitulés sur les axes
plt.xlabel("Tranche d'âge")
plt.ylabel('Nombre de réponses')

plt.show()
```
